```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import requests
from bs4 import BeautifulSoup
```

Since our provided dataset includes movies from 2010 - 2018, we need to scrape movies info from the-numbers.com from 2010 - 2018 as well. Since the website only let us get 100 movies at a time, we should get movie report of each year separatedly so that we can get as many movies as we can and then merge all together.

Let's start with top 100 grossing movies in 2010 from the-numbers.com

```python
#specify URL
url = 'https://www.the-numbers.com/movies/report/All/All/All/All/All/All/All/All/All/None/None/2010/2010/None/None/None/None/None/None?show-release-date=On&view-order-by=domestic-box-office&show-release-year=On&view-order-direction=desc&show-production-budget=On&show-domestic-box-office=On&show-international-box-office=On&show-worldwide-box-office=On&show-genre=On&show-source=On&show-production-method=On&show-creative-type=On'

#make a GET request to fetch the raw HTML content
response = requests.get(url).text

#parse the html content
soup = BeautifulSoup(response, 'lxml')
```

We observe that to get the next dataframe of the next year, the url only changes where the year number changes. We can write a function to do this for us.

```python
def get_table(year):
    url = f'https://www.the-numbers.com/movies/report/All/All/All/All/All/All/All/All/All/None/None/{year}/{year}/None/None/None/None/None/None?show-release-date=On&view-order-by=domestic-box-office&show-release-year=On&view-order-direction=desc&show-production-budget=On&show-domestic-box-office=On&show-international-box-office=On&show-worldwide-box-office=On&show-genre=On&show-source=On&show-production-method=On&show-creative-type=On'
    #make a GET request to fetch the raw HTML content
    response = requests.get(url).text

    #parse the html content
    soup = BeautifulSoup(response, 'lxml')

    #get the table
    table = soup.find('table')

    #read html table into dataframe
    df = pd.read_html(str(table))[0]
    return df
```

```python
df_2012 = get_table(2012)
df_2012.head(2)
```

```python
df_2013 = get_table(2013)
df_2013.head(2)
```

```python
df_2014 = get_table(2014)
df_2014.head(2)
```

```python
df_2015 = get_table(2015)
df_2015.head(2)
```

```python
df_2016 = get_table(2016)
df_2016.head(2)
```

```python
df_2017 = get_table(2017)
df_2017.head(2)
```

```python
df_2018 = get_table(2018)
df_2018.head(2)
```

## II.2.3 Data Cleaning & Preparation

Merge all together

```python
#merge all tables together
merged_df_2 = pd.concat([df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018], ignore_index=True)
```

```python
#preview
merged_df_2.head()
```

```python
#get data summary
merged_df_2.info()
```

```python
#check for null values
merged_df_2.isnull().sum()
```

```python
#check for duplicates
merged_df_2.duplicated().sum()
```

```python
#drop unnecessary column
merged_df_2 = merged_df_2.drop('Unnamed: 0', axis=1)
```

```python
#change column names
merged_df_2 = merged_df_2.rename(columns={'Released':'release_date', 'Released.1':'release_year', 
                                          'Title':'movie', 'Genre':'genre', 'Source':'source','ProductionMethod':'production_method',
                                         'CreativeType':'creative_type', 'ProductionBudget':'production_budget',
                                         'DomesticBox Office':'domestic_gross', 'InternationalBox Office':'international_gross',
                                         'WorldwideBox Office':'worldwide_gross'})
```

```python
#review
merged_df_2.head(2)
```

```python
#check to see if release_date is datetime object
import datetime
isinstance(merged_df_2['release_date'][0], datetime.date)
```

```python
#convert to datetime object
merged_df_2['release_date'] = pd.to_datetime(merged_df_2['release_date'])
```

```python
#check again
isinstance(merged_df_2['release_date'][0], datetime.date)
```

```python
#change release_year into integer
merged_df_2['release_year'].astype(int).dtype
```

```python
#delete '$' and ',' and convert to int
def dollar_clean(dataframe, col):
    dataframe[col] = dataframe[col].str.replace('$', '').str.replace(',', '').astype('int')
    return dataframe
```

```python
dollar_clean(merged_df_2, 'production_budget')
dollar_clean(merged_df_2, 'domestic_gross')
dollar_clean(merged_df_2, 'international_gross')
dollar_clean(merged_df_2, 'worldwide_gross')
```

```python
#compute profit
merged_df_2['worldwide_profit'] = merged_df_2.apply(lambda x: x['worldwide_gross'] - x['production_budget'], axis = 1)
merged_df_2['domestic_profit'] = merged_df_2.apply(lambda x: x['domestic_gross'] - x['production_budget'], axis = 1)
merged_df_2['international_profit'] = merged_df_2.apply(lambda x: x['international_gross'] - x['production_budget'], axis = 1)
```

```python
#compute worldwide_profit_margin column
merged_df_2['worldwide_profit_margin'] = merged_df_2.apply(lambda x: ((x['worldwide_profit']) / (x['worldwide_gross']))*100, axis = 1)
```

```python
#split budgets into bins
bins = [0, 50000000, 100000000, 150000000, 200000000, np.inf]
bin_labels = ['<50M', '50-100M', '100-150M','150-200M', '>200M']

#add 'budget_range' column
merged_df_2['budget_range'] = pd.cut(merged_df_2['production_budget'], bins, labels = bin_labels)
merged_df_2.head(2)
```

```python
#check bins
merged_df_2['budget_range'].value_counts()
```

```python
#save
merged_df_2.to_csv('merged_df_2')
```
